# 2. Primative Types and Variables

## Character Types
A **single character** (*16-bit* number of a Unicode table character).   
It is declared in C# with the keyword `char`.  
  
![you fucked up](../_img/CSharp_CharTypes.jpg)

---

### `char`:
**Letters** or **Other Characters** which are enclosed in **apostrophes** `'.'`.
  

```
for(char c = 'A'; c <= 'z'; c++) 
{

    if(c > 'Z' && c < 'a') continue;

    Console.Write(
        c
    );

}
```

```
// Min Value
(int)char.MinValue
```

```
// Max Value
(int)char.MaxValue
```
